## Matplotlib basics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

```python
_ = plt.plot([0, 1], [0, 3])
```

```python
plt.plot([0, 1, 2], [0, 3, 2], label='line', color='green')
plt.xlabel('Oś x')
plt.ylabel('Oś y')
plt.title('Wykres')
_ = plt.legend()
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
plt.plot(x, x_cos)
```

### Multi ploting

```python
plt.subplot(121)
plt.plot([0,1,2],[3,2,5])
plt.subplot(122)
plt.plot([0,1,2],[3,1,4], 'r--')
```

```python
fig = plt.figure(figsize=[12,8])

plt.subplot(131)
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[3,1,4], 'r--')
plt.subplot(133)
plt.plot([0,1,2],[3,1,4], 'ro')

fig.show()
```

### Bar charts

```python
plt.bar(x=[0,1,2,3], height=[1.2, 3.2, 1.2, 5.2])
```

```python
plt.bar(x=['a','b','c','d'], height=[1.2, 3.2, 1.2, 5.2])
```

```python
plt.barh(y=['a','b','c','d'], width=[1.2, 3.2, 1.2, 5.2])
```

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)


plt.scatter(x1,x2)
```

```python
x3 = np.random.randn(300) * 50
plt.scatter(x1,x2, s=x3)
```

```python
plt.scatter(x1, x2, s=x3, alpha = 0.5)
```

### Subplots and images

```python
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('office.jpg')
img
```

```python
img.shape
```

```python
plt.imshow(img)
_ = plt.axis('off')
```

```python
fig = plt.figure(figsize=(12,10))
plt.subplot(221)
plt.imshow(mpimg.imread('office.jpg'))
plt.axis('off')
plt.subplot(222)
plt.imshow(mpimg.imread('business.jpg'))
plt.axis('off')
plt.subplot(223)
plt.imshow(mpimg.imread('children.jpg'))
plt.axis('off')
plt.subplot(224)
plt.imshow(mpimg.imread('workplace.jpg'))
_ = plt.axis('off')
```

```python
fig = plt.figure(figsize=(12,10))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
    plt.subplot(141 + idx)
    plt.imshow(mpimg.imread(img))
    plt.axis('off')
```
